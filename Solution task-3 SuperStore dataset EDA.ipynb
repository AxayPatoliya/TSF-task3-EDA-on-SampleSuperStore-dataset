## Importing required libraries

```python
import pandas as pd
import seaborn as sns
```

## Loading the dataset

```python
df = pd.read_csv("SampleSuperstore.csv")
```

```python
df.head(10)
```

## Checking null values in the dataset

```python
df.isnull()
```

## Taking a look on visual representation of null values

```python
sns.heatmap(df.isnull(), yticklabels=False, cbar=False, cmap='viridis')
```

### =>As we can see there is no null values in the dataset. So we can proceed further!<=

## Checking different values present in the column named "Segment"

```python
print(df['Segment'].unique())
```

## As we can see, there are 3 categorical values in the 'Segment' column
## We can convert them into integer values by creating the dummies of their

```python
pd.get_dummies(df['Segment'], drop_first=True)
```

```python
# Crearing dummies

Corporate = pd.get_dummies(df['Segment'], drop_first=True)
```

```python
# Dropping the Segment column

df.drop(['Segment'], axis=1, inplace=True)
```

```python
df.head(10)
```

```python
# Adding the created integer column

df = pd.concat([df, Corporate], axis=1)
```

```python
df.head(10)
```

Now, if 'Corporate' column contains 1 that means Segment is corporate, if 'Home Office' contains 1 that means Segment is Home Office and if both are 0 then Segment contains Consumer.

## Checking different values present in the column named "Category"

```python
print(df['Category'].unique())
```

### We are gonna perform the same operation we'd done for "Segment" column

### There are 3 values in "Category" column and they are in categorical manner, we can convert them into integer by creating the dummies of theirs.

```python
# Crearing dummies

Category = pd.get_dummies(df['Category'], drop_first=True)
```

```python
# Dropping the 'Category' column

df.drop(['Category'], axis=1, inplace=True)
```

```python
df.head(10)
```

```python
# Adding the created integer column

df = pd.concat([df, Category], axis=1)
```

```python
df.head(10)
```

If 'Office Supplies' column contains 1 that means category is office supplies, if 'Technology' contains 1 that means category is technology if both are 0 then category is furniture.

## Checking the unique values in the 'Country' column

```python
print(df['Country'].unique())
```

As we can see there is only one value in the column

## Dropping out the 'Country' column as it doesn't affects the sales'' as much

```python
df.drop(['Country'], axis=1, inplace=True)
```

```python
df.head(5)
```

### Now, data looks quite clear and we will see the corelation of 'Sales' column with others

```python
df[df.columns[1:]].corr()['Sales'][:]
```
